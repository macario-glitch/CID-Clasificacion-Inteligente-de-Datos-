# ***Fundamentos de la técnica (Regresión logística)***

La Regresión lógistca estima la probabilidad de que ocurra un evento, (0-1, Verdadero-Falso, etc.) en función de un conjunto de datos determinado de variables independientes.

**Ejemplo:** Si el modelo dice ".87", entonces significa que hay un 87% de que sea clase 1. Se aplica un umbral: Si probabilidad >= 0.5 es 1, si probabilidad < 0.5 es 0.

**Funcionamiento:**


1.   El modelo combina las variables con una ecuacion lineal. (z=b0 + b1x...)
2.   Se le aplica la función sigmoide (σ(z))
3.   Se obtiene una probabilidad entre 0 y 1
4.   Se decide si es 0 o 1 utilizando un umbral (normalmente de 0.5)


**¿Como es que el modelo aprende?**
Para ajustar los parametros (*b*) se puede utilizar:


*   **Función de Costo:** Mide que tan mal predice el modelo. Entre más bajo, mejor.
*   **Gradiente descendiente:** Un algoritmo que va ajustando los parametros para minimizar el costo.

***Preprocesamiento***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer #Dataset
from sklearn.model_selection import train_test_split #Entrenamiento/Prueba
from sklearn.preprocessing import StandardScaler #Normalizar
from sklearn.linear_model import LogisticRegression #Modelo
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay #Evaluación

cancer = load_breast_cancer()
X = pd.DataFrame(cancer.data, columns=cancer.feature_names)
y = pd.DataFrame(cancer.target, columns=["target"])

print("Primeras tuplas:")
display(X.head())
print("\nAnálisis estadistico:")
display(X.describe())

df = pd.concat([X, y], axis=1)
cols = ['mean compactness', 'mean concavity', 'mean concave points', 'mean symmetry', 'mean fractal dimension', 'target']

sns.pairplot(df[cols], hue="target", palette="Set2", diag_kind="hist")
plt.suptitle("Gráfico de dispersión (solo 5 variables) - Breast Cancer", y=1.02)
plt.show()

#División de datos (Entrenamiento/Prueba)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

#Normalizacion
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

***Entrenamiento***

```python
model = LogisticRegression(max_iter=500, random_state=42)
model.fit(X_train_scaled, y_train)
```

***Predicción***

```python
def predecir_clase(patron, valor):
    #Recibe un patrón. Devuelve la clase predicha por el modelo.
    patron = np.array(patron).reshape(1, -1)
    patron_scaled = scaler.transform(patron)
    prediccion = model.predict(patron_scaled)[0]

    clase = "Maligno (1)" if prediccion == 1 else "Benigno (0)"
    print("Clase predicha:", clase, "Clase esperada:", valor[0], "\n")


# Ejemplo de uso: (tomamos un registro real)
ejemplo = X.iloc[250].values
valor_esperado = y.iloc[250]
print("\nPredicción de ejemplo con un patrón random del dataset:")
predecir_clase(ejemplo,valor_esperado)
```

***Evaluación***

```python
y_pred = model.predict(X_test_scaled)
mat_conf = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(mat_conf, annot=True, fmt="d", cmap="Blues")
plt.title("Matriz de Confusion - Regresión Logística")
plt.xlabel("Predicción")
plt.ylabel("Valor Real")
plt.show()

accuracy = accuracy_score(y_test, y_pred)
print(f"\nEXACTITUD DEL MODELO\n{accuracy:.4f}")
```
