# Fundamento de la técnica: **Kmeans**

K-means es una técnica de agrupamiento no supervisado (clustering) utilizada para dividir un conjunto de datos en **k** grupos similares entre sí, sin que el algoritmo sepa de antemano a qué grupo pertenece cada dato.

El objetivo del algoritmo es encontrar K grupos dentro de los datos, de modo que:


*   Los datos dentro del mismo grupo sean lo más parecidos posibles.
*   Los grupos diferentes sean lo más distintos posible entre sí.

**Funcionamiento**

**Elegir K:** El numero de grupos (K) se define de antemano. Aun que la elección puede ser arbitriaria, la mejor forma de decidir el número optimo de K es con el metodo del codo (Elbow Method):

El cual consiste en ejecutar K-means con distintos valores de k (Minimo: 1 al 10). Luego se calcula el ***error cuadrático total*** (SSE) para cada valor de K. Para finalmente hacer una grafica y comparar los resultados. El punto donde el error deja de disminuir bruscamente (Forma un codo) indica el mejor valor de K.

**Inicializar centroide:** El algoritmo elige K puntos iniciales aleatorios que representan los centros de los grupos (centroides).

**Asignar cada dato a su grupo más cercano:** Para cada punto, se calcula la distancia al centroide más cercano (Usualmente se usa la distancia Euclidiana) y se le asigna al grupo correspondiente.

**Recalcular los centroides:** Una vez que cada punto ha sido asignado a un grupo, se recalcula la posición de los centroides como el promedio de todos los puntos que pertenecen a cada grupo.

**Repetir:** Los pasos 3 y 4 se repiten hasta que:


*   Los centroides ya no cambian mucho, o...
*   Se alcanza el número maximo de iteraciones definido.

La cual:

**xi:** Vector de caracteristicas del dato i. [xi1,...,xin]

**μj:** Centroide (promedio) del grupo j.

**∥xi​−μj​∥2:** Distancia cuadrada (normalmente euclidiana) entre el punto *xi* y el centroide *μj*.

**J:** Error total o SSE (Sum of Squared Errors), lo que el algoritmo intenta minimizar.

Se repite hasta que *J* (La función de costo) ya no cambie significativamente.

**Procesamiento**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

cancer = load_breast_cancer()
X = pd.DataFrame(cancer.data, columns=cancer.feature_names)
y = pd.DataFrame(cancer.target, columns=["target"]) #Solo para evaluar

print("Primeras tuplas:")
display(X.head())
print("\nAnálisis estadistico:")
display(X.describe())

#Normalizar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

df = pd.concat([X, y], axis=1)
cols = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness', 'target']

sns.pairplot(df[cols], hue="target", palette="Set2", diag_kind="hist")
plt.suptitle("Gráfico de dispersión (solo 5 variables) - Breast Cancer", y=1.02)
plt.show()
```

```python
#Metodo del Codo. Ya se sabe que este dataset tiene 2 clases, pero vamos a ver como funciona el metodo del codo.
inercia = []  #Lista para guardar la inercia (SSE) para cada valor de K
K_range = range(1, 11)  #Probaremos de K=1 a K=10

for k in K_range:
    kmeans_temp = KMeans(n_clusters=k, random_state=5)
    kmeans_temp.fit(X_scaled)
    inercia.append(kmeans_temp.inertia_)  #Inercia: suma de distancias cuadradas al centroide más cercano

#Graficar el resultado
plt.figure(figsize=(8,6))
plt.plot(K_range, inercia, 'bo-', markersize=8)
plt.title('Método del Codo (Elbow Method)')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('Inercia (Suma de Distancias Cuadradas)')
plt.grid(True)
plt.show()

kmeans = KMeans(n_clusters=2, random_state=5)
kmeans.fit(X_scaled)
```

**Prediction**

```python
def predecir_patron(patron, resultado):
    """
    Recibe un vector (lista o array) con las mismas características
    que el dataset original y predice a qué cluster pertenece.
    """
    patron = np.array(patron).reshape(1, -1)
    patron_scaled = scaler.transform(patron)
    cluster_predicho = kmeans.predict(patron_scaled)[0]
    print(f"\nEl patrón ingresado pertenece al cluster: {cluster_predicho}\n")
    print(f"Resultado esperado: {resultado}\n")
    return cluster_predicho

# Ejemplo de uso:
ejemplo = X.iloc[34].values
resultado = y.iloc[34].values
predecir_patron(ejemplo, resultado)
```

**Model Evaluation**

```python
# Predicciones del modelo
y_pred = kmeans.labels_

#Matriz de confusión
mat_conf = confusion_matrix(y, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(mat_conf, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Benigno", "Maligno"],
            yticklabels=["Benigno", "Maligno"])
plt.title("Matriz de Confusión (K-Means vs. Etiquetas reales)")
plt.xlabel("Clusters Reales")
plt.ylabel("Clases Predichos")
plt.show()

#Exactitud
accuracy = accuracy_score(y, y_pred)
print(f"\nExactitud (Accuracy) del modelo: {accuracy:.3f}")
```
